```python
# remove nan, round columns and remove first column
spotify = spotify.dropna()
columns_to_round = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
spotify[columns_to_round] = spotify[columns_to_round].round(2)
spotify = spotify.drop(spotify.columns[0], axis=1)
spotify.head()
```

```python
# delete copies (keep the most popular one)
spotify2 = spotify.loc[spotify.groupby(['duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode',
                                     'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
                                     'time_signature', 'track_genre'])['popularity'].idxmax()]
```

```python
spotify2.head()
```

```python
spotify2.describe()
```

```python
len(spotify2)
```

```python
# delete unpopular songs (unfortunately there still were copies)
spotify2 = spotify2[spotify2['popularity'] >= 10]
```

```python
len(spotify2)
```

```python
spotify2['track_id'].nunique()
```

```python
genre_count = spotify2['track_genre'].value_counts()
genre = genre_count.to_dict()
genre
```

```python
spotify2['genre_count'] = spotify2['track_genre'].map(genre)
spotify2.tail()
```

```python
# for multi-genre (and then multi-record) songs, keep only the most frequent genre
spotify3 = spotify2.loc[spotify2.groupby(['track_id'])['genre_count'].idxmax()]
spotify3['explicit'] = spotify3['explicit'].astype(int)
spotify3.head()
```

```python
spotify3.describe()
```

```python
spotify3['track_genre'].value_counts()
```

```python
# create df with only numerical features (binary explicit and mode included)
spotify_numeric = spotify3.loc[:, ['duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 
                                   'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 
                                   'valence', 'tempo', 'time_signature', 'popularity']].reset_index(drop=True)

spotify_numeric.head()
```

```python
spotify_numeric['popularity'].plot(kind='density', linewidth=2)  
plt.xlabel('Popularity')  
plt.ylabel('Density')  
plt.xlim((10, 100))
plt.ylim((0, 0.035))

plt.show()
```

```python
# create column for album artist (necessary to avoid problems from album homonymy)
# the first name in the 'artists' column always seems the album artist so this rule has been used
spotify3['album_artist'] = spotify3['artists'].str.split(';', n=1).str[0]
spotify3.head()
```

```python
artists_set = set()

for value in spotify3['artists']:
    parts = value.split(';')
    artists_set.update(parts)

artists_set
```

```python
artist_popularity_dict = {}

for index, row in spotify3.iterrows():
    artists_list = row['artists'].split(';')
    
    popularity = row['popularity']
    
    for artist in artists_list:
        if artist in artist_popularity_dict:
            artist_popularity_dict[artist]['sum'] += popularity
            artist_popularity_dict[artist]['count'] += 1
        else:
            artist_popularity_dict[artist] = {'sum': popularity, 'count': 1}

for artist, data in artist_popularity_dict.items():
    avg_popularity = data['sum'] / data['count']
    artist_popularity_dict[artist] = avg_popularity

artist_popularity_dict
```

```python
# created column 'main_artist' with the most popular artist present in the song
# artist popularity computed as the mean of the popularity of all the songs where that artist is present
# artist popularity added as a column as well
spotify3['main_artist'] = spotify3['artists'].apply(lambda x: max(x.split(';'), key=lambda artist: artist_popularity_dict.get(artist, 0)))
spotify3['artist_popularity'] = spotify3['main_artist'].map(artist_popularity_dict).round(2)
```

```python
spotify3.head()
```

```python
# created column for genre popularity, same computations as for the artists
spotify3['genre_popularity'] = spotify3.groupby('track_genre')['popularity'].transform('mean').round(2)
```

```python
spotify3.head()
```

```python
# created column for album popularity, same computations as for the artists and genres
# two songs are considered part of the same album if 'album_name' and 'album_artist' are the sames for both songs
spotify3['album_popularity'] = spotify3.groupby(['album_name', 'album_artist'])['popularity'].transform('mean').round(2)
```

```python
spotify3.head()
```

```python
# created df with all the features
spotify_complete = spotify3.loc[:, ['duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                                    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 
                                    'album_popularity', 'artist_popularity', 'genre_popularity', 'popularity']].reset_index(drop=True)

spotify_complete.head()
```

```python
spotify_numeric.to_csv('data/numeric.csv', index=False) 
spotify_complete.to_csv('data/complete.csv', index=False) 
```
